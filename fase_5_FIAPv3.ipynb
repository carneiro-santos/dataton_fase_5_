```python

# 1. DEFINA A FUNÇÃO QUE ESTÁ FALTANDO
def extrair_fase(valor):
    if pd.isna(valor):
        return np.nan
    v_str = str(valor).strip().upper()
    if v_str == 'ALFA':
        return 0
    try:
        # Trata números decimais em string (ex: '1.0' ou '2,0')
        return int(float(v_str.replace(',', '.')))
    except:
        return np.nan

# 2. SEÇÃO DE TRATAMENTO (Seu código original com correções)
# 4a. Tratamento INDE 2024
if "inde_ano" in df_2024.columns:
    df_2024["inde_ano"] = pd.to_numeric(df_2024["inde_ano"].replace("INCLUIR", np.nan), errors="coerce")

# 4b. Fases
df_2022["fase"] = pd.to_numeric(df_2022["fase"], errors="coerce")
df_2023["fase"] = df_2023["fase"].apply(extrair_fase)
df_2024["fase"] = df_2024["fase"].apply(extrair_fase)

# 4c. Datas de Nascimento
for df_temp, label in [(df_2022, "2022"), (df_2023, "2023"), (df_2024, "2024")]:
    if "data_nasc" in df_temp.columns:
        if label == "2022" and df_temp["data_nasc"].astype(str).str.len().max() <= 4:
            df_temp["data_nasc"] = pd.to_datetime(df_temp["data_nasc"].astype(str) + "-01-01", errors="coerce")
        else:
            df_temp["data_nasc"] = pd.to_datetime(df_temp["data_nasc"], errors="coerce")

# 4d. Colunas Numéricas
NUMERIC_COLS = [
    "inde_ano", "inde_2022", "inde_2023", "pedra_ano",
    "iaa", "ieg", "ips", "ipp", "ida", "ipv", "ian",
    "nota_matematica", "nota_portugues", "nota_ingles",
    "num_avaliadores", "defasagem", "idade", "ano_ingresso",
]

for df_loop in [df_2022, df_2023, df_2024]:
    for col in NUMERIC_COLS:
        if col in df_loop.columns:
            df_loop[col] = pd.to_numeric(df_loop[col], errors="coerce")

```

```python

# 1. Definição da função de engenharia reversa
def calcular_ipp_reverso(row):
    """
    Tenta inferir o IPP a partir do INDE e outros indicadores conhecidos.
    Fórmula base: INDE = (IAA*0.1 + IEG*0.2 + IPS*0.1 + IPP*0.1 + IDA*0.2 + IPV*0.2 + IAN*0.1)
    """
    try:
        # Se já tiver IPP ou não tiver INDE, não calcula
        if pd.notna(row.get('ipp')) or pd.isna(row.get('inde_ano')):
            return row.get('ipp')

        # Soma dos outros componentes multiplicados por seus pesos
        outros = (
            (row.get('iaa', 0) * 0.1) +
            (row.get('ieg', 0) * 0.2) +
            (row.get('ips', 0) * 0.1) +
            (row.get('ida', 0) * 0.2) +
            (row.get('ipv', 0) * 0.2) +
            (row.get('ian', 0) * 0.1)
        )

        # O que sobra do INDE deve ser o (IPP * 0.1)
        ipp_inferido = (row['inde_ano'] - outros) / 0.1
        return round(max(0, min(10, ipp_inferido)), 3) # Garante que fique entre 0 e 10
    except:
        return np.nan

# ---------------------------------------------------------------------------
# 4.5 INFERÊNCIA DO IPP POR ENGENHARIA REVERSA (Nomes Corrigidos)
# ---------------------------------------------------------------------------
secao("4.5 INFERINDO IPP POR ENGENHARIA REVERSA")

# Ajustando para df_2022
if "ipp" not in df_2022.columns:
    df_2022["ipp"] = np.nan

mask_sem_ipp = df_2022["ipp"].isna()
df_2022.loc[mask_sem_ipp, "ipp"] = df_2022[mask_sem_ipp].apply(calcular_ipp_reverso, axis=1)

print(f"  [2022] IPP inferido para {df_2022['ipp'].notna().sum()} alunos.")
print(f"  [2022] IPP ainda nulo: {df_2022['ipp'].isna().sum()}")

# Ajustando para df_2023 e df_2024
for df_loop, label in [(df_2023, "2023"), (df_2024, "2024")]:
    if "ipp" in df_loop.columns:
        mask = df_loop["ipp"].isna()
        n_antes = mask.sum()
        if n_antes > 0:
            df_loop.loc[mask, "ipp"] = df_loop[mask].apply(calcular_ipp_reverso, axis=1)
            n_depois = df_loop["ipp"].isna().sum()
            print(f"  [{label}] {n_antes} IPP nulos → {n_antes - n_depois} inferidos")
```

```python
# 1. Definição da função auxiliar necessária
def normalizar_instituicao(inst):
    if pd.isna(inst):
        return "Não Informado"
    inst = str(inst).upper()
    if "ESCOL" in inst or "PUBLI" in inst:
        return "Escola Pública"
    if "PARTIC" in inst or "PRIVAD" in inst:
        return "Escola Particular"
    return "Outros"

# ---------------------------------------------------------------------------
# 5. PADRONIZAÇÃO DE CATEGORIAS (Nomes Corrigidos)
# ---------------------------------------------------------------------------
secao("5. PADRONIZANDO CATEGORIAS")

GENERO_MAP = {
    "menina": "Feminino", "menino": "Masculino",
    "feminino": "Feminino", "masculino": "Masculino",
}

# Usando os nomes corretos: df_2022, df_2023, df_2024
for df_loop in [df_2022, df_2023, df_2024]:
    if "genero" in df_loop.columns:
        df_loop["genero"] = df_loop["genero"].str.strip().str.lower().map(GENERO_MAP)
print("  [Todas] genero: padronizado → 'Feminino' / 'Masculino'")

for df_loop in [df_2022, df_2023, df_2024]:
    if "instituicao" in df_loop.columns:
        df_loop["instituicao"] = df_loop["instituicao"].apply(normalizar_instituicao)
print("  [Todas] instituicao: categorias normalizadas")

PEDRA_ORDER = {"Quartzo": 1, "Ágata": 2, "Ametista": 3, "Topázio": 4}
PEDRA_COLS = ["pedra_ano", "pedra_2020", "pedra_2021", "pedra_2022", "pedra_2023"]

for df_loop in [df_2022, df_2023, df_2024]:
    for col in PEDRA_COLS:
        if col in df_loop.columns:
            df_loop[col] = df_loop[col].map(PEDRA_ORDER)
print("  [Todas] pedra_*: Quartzo=1, Ágata=2, Ametista=3, Topázio=4")
```

```python
# 1. Definição da função de imputação necessária
def impute_by_group(df, col, strategy="median"):
    """
    Preenche valores nulos usando a estratégia escolhida (mediana)
    agrupada pela 'fase' do aluno.
    """
    if strategy == "median":
        # Preenche com a mediana da fase
        df[col] = df[col].fillna(df.groupby('fase')[col].transform('median'))
        # Se ainda sobrarem nulos (fases com apenas 1 aluno nulo), usa a mediana global
        df[col] = df[col].fillna(df[col].median())
    return df

# 2. Configuração de colunas (caso não tenham sido definidas na célula anterior)
INDICATOR_COLS = ["iaa", "ieg", "ips", "ipp", "ida", "ipv", "ian"]
NOTE_COLS      = ["nota_matematica", "nota_portugues"]
PEDRA_COLS     = ["pedra_ano", "pedra_2020", "pedra_2021", "pedra_2022", "pedra_2023"]

# ---------------------------------------------------------------------------
# 6. TRATAMENTO DE VALORES NULOS (Nomes Corrigidos)
# ---------------------------------------------------------------------------
secao("6. TRATANDO VALORES NULOS")

# Usando df_2022, df_2023, df_2024
dataframes_list = [(df_2022, "2022"), (df_2023, "2023"), (df_2024, "2024")]

for i, (df_loop, label) in enumerate(dataframes_list):
    # Indicadores e Notas Principais
    for col in INDICATOR_COLS + NOTE_COLS:
        if col in df_loop.columns and df_loop[col].isna().any():
            antes = df_loop[col].isna().sum()
            df_loop = impute_by_group(df_loop, col, strategy="median")
            depois = df_loop[col].isna().sum()
            print(f"  [{label}] {col}: {antes} nulos → {depois} após imputação")

    # Inglês: ausência estrutural → 0
    if "nota_ingles" in df_loop.columns:
        antes = df_loop["nota_ingles"].isna().sum()
        df_loop["nota_ingles"] = df_loop["nota_ingles"].fillna(0)
        if antes > 0:
            print(f"  [{label}] nota_ingles: {antes} nulos → preenchidos com 0")

    # Pedras históricas: -1 = sem histórico
    for col in PEDRA_COLS:
        if col in df_loop.columns:
            df_loop[col] = df_loop[col].fillna(-1)

    # Idade
    if "idade" in df_loop.columns and df_loop["idade"].isna().any():
        df_loop = impute_by_group(df_loop, "idade", strategy="median")

    # Defasagem
    if "defasagem" in df_loop.columns:
        df_loop["defasagem"] = df_loop["defasagem"].fillna(0)

    # INDE histórico
    for col in ["inde_2022", "inde_2023"]:
        if col in df_loop.columns:
            df_loop[col] = df_loop[col].fillna(-1)

    # INDE do ano corrente
    if "inde_ano" in df_loop.columns and df_loop["inde_ano"].isna().any():
        antes = df_loop["inde_ano"].isna().sum()
        df_loop = impute_by_group(df_loop, "inde_ano", strategy="median")
        print(f"  [{label}] inde_ano: {antes} nulos → imputados pela mediana da fase")

    # Reatribuição manual para garantir que os DFs originais sejam atualizados
    if label == "2022": df_2022 = df_loop
    elif label == "2023": df_2023 = df_loop
    elif label == "2024": df_2024 = df_loop

print("\nTratamento de nulos finalizado com sucesso.")
```

```python
# 1. Definição da função de Engenharia de Features
def add_features(df, ano_atual):
    # a. Anos no programa
    if "ano_ingresso" in df.columns:
        df["anos_no_programa"] = ano_atual - df["ano_ingresso"]

    # b. Média de Notas (Matemática e Português)
    notas = ["nota_matematica", "nota_portugues"]
    df["media_notas"] = df[notas].mean(axis=1)

    # c. Média de Indicadores (IAA, IEG, IPS, IPP, IDA, IPV, IAN)
    indicadores = ["iaa", "ieg", "ips", "ipp", "ida", "ipv", "ian"]
    cols_presentes = [c for c in indicadores if c in df.columns]
    df["media_indicadores"] = df[cols_presentes].mean(axis=1)

    # d. Evolução de Pedra (Comparação com o ano anterior, se existir)
    # Pedra_ano vs Pedra_(ano_anterior)
    col_pedra_prev = f"pedra_{ano_atual - 1}"
    if "pedra_ano" in df.columns and col_pedra_prev in df.columns:
        # 0 = manteve, positivo = subiu, negativo = desceu
        df["evolucao_pedra"] = df["pedra_ano"] - df[col_pedra_prev]
    else:
        df["evolucao_pedra"] = 0

    # e. Codificação Categórica (Dummies/Manual)
    if "genero" in df.columns:
        df["genero_feminino"] = (df["genero"] == "Feminino").astype(int)

    if "instituicao" in df.columns:
        # Transforma instituição em código numérico para o modelo
        df["instituicao_cod"] = df["instituicao"].astype("category").cat.codes

    return df

# ---------------------------------------------------------------------------
# 7. ENGENHARIA DE FEATURES (Nomes Corrigidos)
# ---------------------------------------------------------------------------
secao("7. ENGENHARIA DE FEATURES")

# Aplicando nos nomes de variáveis df_2022, df_2023, df_2024
df_2022 = add_features(df_2022, 2022)
df_2023 = add_features(df_2023, 2023)
df_2024 = add_features(df_2024, 2024)

print("  Novas colunas criadas: anos_no_programa, media_notas, media_indicadores,")
print("  evolucao_pedra, genero_feminino, instituicao_cod")
```

```python
# 1. Definição da função safe_select
def safe_select(df, cols):
    """
    Seleciona colunas que existem no DataFrame, evitando erros de colunas ausentes.
    """
    cols_presentes = [c for c in cols if c in df.columns]
    return df[cols_presentes].copy()

# ---------------------------------------------------------------------------
# 8. CONSOLIDAÇÃO (Nomes Corrigidos)
# ---------------------------------------------------------------------------
secao("8. CONSOLIDANDO BASE ÚNICA")

CORE_COLS = [
    "ra", "ano_referencia", "fase", "genero", "genero_feminino",
    "idade", "data_nasc", "ano_ingresso", "anos_no_programa",
    "instituicao", "instituicao_cod",
    "pedra_ano", "pedra_2020", "pedra_2021", "pedra_2022",
    "inde_ano", "inde_2022",
    "num_avaliadores",
    "iaa", "ieg", "ips", "ipp", "ida", "ipv", "ian",
    "nota_matematica", "nota_portugues", "nota_ingles",
    "media_notas", "media_indicadores", "evolucao_pedra",
    "defasagem",
]

# Seleção segura usando os nomes de variáveis corretos
base_22 = safe_select(df_2022, CORE_COLS)
base_23 = safe_select(df_2023, CORE_COLS + ["pedra_2023", "inde_2023"])
base_24 = safe_select(df_2024, CORE_COLS + ["pedra_2023", "inde_2023"])

# Concatenação final
base_consolidada = pd.concat([base_22, base_23, base_24], ignore_index=True)

print(f"  Base consolidada: {base_consolidada.shape[0]} linhas | {base_consolidada.shape[1]} colunas")

# Verificação de nulos
nulos_finais = base_consolidada.isna().sum()
nulos_finais = nulos_finais[nulos_finais > 0]

if len(nulos_finais) > 0:
    print("\n  Nulos remanescentes (ausência estrutural entre anos):")
    for col, n in nulos_finais.items():
        print(f"    {col}: {n} ({n/len(base_consolidada)*100:.1f}%)")
else:
    print("  Sem nulos remanescentes nas colunas core.")
```
